```python
from tensorflow.keras.layers import Activation,BatchNormalization,Conv2D,Conv2DTranspose,MaxPooling2D,Concatenate,Input
from tensorflow.keras.models import Model
```

```python
def conv_2d(input,no_of_features):
    x=Conv2D(no_of_features,padding='same',kernel_size=(3,3)) (input)
    x=BatchNormalization()(x)
    x=Activation('relu')(x)
    x=Conv2D(no_of_features,padding='same',kernel_size=(3,3)) (input)
    x=BatchNormalization()(x)
    x=Activation('relu')(x)
    return x
```

```python
def encoder(input,no_of_features):
    x=conv_2d(input,no_of_features)
    p=MaxPooling2D((2,2))(x)
    return x,p
```

```python
def decoder(input,no_of_features,skip_connection):
    x=Conv2DTranspose(no_of_features,strides=2,kernel_size=2,padding='same')(input)
    x=Concatenate()([x,skip_connection])
    x=conv_2d(x,no_of_features)
    return x
```

```python
def bulid_unet(input_shape):
    input=Input(input_shape)
    s1,p1=encoder(input,64)
    s2,p2=encoder(p1,128)
    s3,p3=encoder(p2,256)
    s4,p4=encoder(p3,512)

    b=conv_2d(p4,1024)

    d1=decoder(b,512,s4)
    d2=decoder(d1,256,s3)
    d3=decoder(d2,128,s2)
    d4=decoder(d3,64,s1)
    # single class segmentation
    output=Conv2D(1,1,padding='same',activation='sigmoid')(d4)

    model=Model(input,output,name='UNET')
    return model
```

```python
model=bulid_unet((256,256,3))
```
